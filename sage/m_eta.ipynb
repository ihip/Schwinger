# $m_\eta$

Mass of $m_\eta$ in the chiral limit is
$$\mu = g \sqrt\frac{N_f}{\pi} = \sqrt\frac{N_f}{\pi\beta}$$

Analytic prediction for $m_\eta$ as a function of $m$ is given in \[Harada1994\] eq. (3.15):
$$\quad \ln(m_\eta - \sqrt{2}) = 0.19 + 0.993 \ln m$$
where $m$ and $m_\eta$ are given in units of $g/\sqrt{\pi}$ (see section III.C.).

Solving for $m_\eta$ gives
$$m_\eta = e^{0.19 + 0.993 \ln m} + \sqrt{2}$$
and simplified version of the above formula is
$$m_\eta = e^{0.19} m^{0.993} + \sqrt{2}$$

To be able to compare with the lattice simulation results and "semi-classical" prediction from \[Gattringer1999\]
$$m_\eta = \sqrt{\mu^2 + [2.1633(m^2/\sqrt{\beta})^{1/3}]^2}$$
one should modifiy Harada's formula so that
$$ m \rightarrow m \sqrt{\pi\beta} \quad \& \quad m_\eta \rightarrow m_\eta \sqrt{\pi\beta}$$
i.e.
$$m_\eta = \frac{e^{0.19} (m \sqrt{\pi\beta})^{0.993} + \sqrt{2}}{\sqrt{\pi\beta}}$$

```python
Nf = 2
β = 4.0
```

```python
def meta(m):
    return (exp(0.19) * (m * sqrt(pi * β))^0.993 + sqrt(2.0)) / sqrt(pi * β)
```

```python
def metasc(m):
    return sqrt(Nf / (pi * β) + (2.1633 * (m^2 / sqrt(β))^(1/3))^2)
```
