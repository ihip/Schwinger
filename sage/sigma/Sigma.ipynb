## Analytical prediction by Hosotani

```python
γ = 0.5772156649 # Euler-Mascheroni constant
```

```python
# input parameters:
β = 4.0
L = 20 # N_t
```

```python
μ = n(sqrt(2 / (pi * β))) # eta-mass
g = 1.0 / sqrt(β)
```

```python
μ
```

```python
g
```

```python
μ * L # this should be >> 1
```

```python
b = 1.0 / (L * sqrt(μ * L)) # solution is valid for m << than this number
```

```python
b
```

```python
k = n(2 * exp(γ) * μ * L / pi^2) # slope for small m
```

```python
k
```

## Simulation results

```python
def readCollectedData2(inFileName):
    f = open(inFileName, "r")
    list2 = []
    for line in f:
        if line[0] == " ":
            data = line.split()
            list2.append([float(data[0]), float(data[1])])
    f.close()
    return list2
```

```python
def divBYg(list, g):
    return [[a[0] / g, a[1] / g] for a in list]
```

```python
Nf0e2 = readCollectedData2("Nf0-100.sig")
```

```python
Nf1e2 = readCollectedData2("Nf1-100.sig")
```

```python
Nf2e2 = readCollectedData2("Nf2-100.sig")
```

```python
Nf0e3 = readCollectedData2("Nf0-1000.sig")
```

```python
Nf1e3 = readCollectedData2("Nf1-1000.sig")
```

```python
Nf2e3 = readCollectedData2("Nf2-1000.sig")
```

```python
var('m')
plot([k * m, (exp(4 * γ) * m * μ^2 / (4 * pi^3))^(1/3)], 0.0001, 0.15, ymax = 0.2,
  gridlines = True, axes_labels=[r'$m$', r'$<\bar{\psi}\psi>$'],
  color = ['blue', 'gray'], linestyle = "--") + line([[b, 0], [b, 0.25]], linestyle = ":",
  color = 'gray') + list_plot(Nf0e3,
  color = 'red') + list_plot(Nf1e3, color = 'black') + list_plot(Nf2e3) + list_plot(Nf0e2,
  color = 'gray') + list_plot(Nf1e2, color = 'gray') + list_plot(Nf2e2, color = 'gray')
```

```python
Nf0e2g = divBYg(Nf0e2, g)
```

```python
Nf1e2g = divBYg(Nf1e2, g)
```

```python
Nf2e2g = divBYg(Nf2e2, g)
```

```python
Nf0e3g = divBYg(Nf0e3, g)
```

```python
var('x') # x = m / g  =>  m = x * g
plot([k * (x * g) / g, (exp(4 * γ) * (x * g) * μ^2 / (4 * pi^3))^(1/3) / g], 0.0001, 0.15 / g,
  gridlines = True, axes_labels=[r'$m/g$', r'$<\bar{\psi}\psi>/g$'], ymax = 0.4,
  color = ['blue', 'gray'], linestyle = "--") + line(
  [[b / g, 0], [b / g, 0.25 / g]], linestyle = ":", color = 'gray') + list_plot(Nf0e3g,
  color = 'red') + list_plot(Nf1e3g, color = 'black') + list_plot(Nf2e3g) + list_plot(Nf0e2g,
  color = 'gray') + list_plot(Nf1e2g, color = 'gray') + list_plot(Nf2e2g, color = 'gray')
```
