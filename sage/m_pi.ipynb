# $m_\pi$

Analytic prediction for $m_\pi$ as a function of $m$ is given in \[Harada1994\] eq. (3.14):

$$\quad \ln m_\pi = 0.564 + 0.5007 \ln m$$

where $m$ and $m_\pi$ are given in units of $g/\sqrt{\pi}$ (see section III.C.).

Solving for $m_\pi$ gives
$$m_\pi = e^{0.564 + 0.5007 \ln m}$$
and simplified version of the above formula is
$$m_\pi = e^{0.564} m^{0.5007}$$

To be able to compare with the lattice simulation results and the predictions from \[Hosotani1995\]

$$m_\pi = 2.1633\,\beta^{-1/6}\,m^{2/3}$$

one should modifiy Harada's formula so that
$$ m \rightarrow m \sqrt{\pi\beta} \quad \& \quad m_\eta \rightarrow m_\eta \sqrt{\pi\beta}$$
i.e.
$$m_\pi = \frac{e^{0.564} (m \sqrt{\pi\beta})^{0.5007}}{\sqrt{\pi\beta}}$$

```python
β = 4.0
```

```python
def mpi(m):
    return (exp(0.564) * (m * sqrt(pi * β))^0.5007) / sqrt(pi * β)
```

```python
def mpiHs(m): # Hosotani prediction
    return 2.1633 * β^(-1/6) * m^(2/3)
```
