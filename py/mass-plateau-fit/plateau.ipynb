```python
import matplotlib.pyplot as plt

list_x = []
list_y = []
list_dy = []

f = open("L20-beta4-nf2-1000.Mj1", "r")
# read line with comment
print(f.readline())

# read line with quark mass, pion mass and error bars
line = f.readline()
data = line.split()
qmass = float(data[0])
pmass = float(data[1])
pmasserr = float(data[2])
print(f'{qmass:.4f}   {pmass:.8f}   {pmasserr:.8f}') 

for line in f:
    if line[0] == " ":
        data = line.split()
        x = float(data[0])
        y = float(data[1])
        dy = float(data[2])
        print(x, y, dy)
        list_x.append(x)
        list_y.append(y)
        list_dy.append(dy)
    if line[0] == "\n":
        plt.title(r"mass plateau for quark mass = " + str(f'{qmass:.4f}'))
        plt.xlabel(r"$t$")
        plt.ylabel(r"$M_\pi$")
        plt.errorbar(list_x[1:], list_y[1:], yerr = list_dy[1:], linestyle = 'None', marker = 'o')
        plt.axhline(pmass, ls = '-', lw = 0.75, c = 'red')
        plt.axhline(pmass - pmasserr, lw = 0.75, ls = ':', c = 'red')
        plt.axhline(pmass + pmasserr, lw = 0.75, ls = ':', c = 'red')
        plt.show()

        # read line with comment
        line = f.readline()
        if(line == ""): break
        print(line, end = '')

        # read line with quark mass, pion mass and error bars
        line = f.readline()
        data = line.split()
        qmass = float(data[0])
        pmass = float(data[1])
        pmasserr = float(data[2])
        print(f'{qmass:.4f}   {pmass:.8f}   {pmasserr:.8f}') 
        
        list_x = []
        list_y = []
        list_dy = []

f.close()
```
